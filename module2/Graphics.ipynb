## Functions : 
* plot, loglog, semilogx, semilogy
* xlabel, ylabel, title
* Latex like
* label and legend
* xlim, ylim
* figure
* subplot
* imshow
* savefig : svg, pdf

```python
import numpy as np


X = np.linspace(0, 10, 200)
X2 = np.linspace(0, 10, 20)

Y = np.sin(X)
Y2 = np.sin(X2)
#plt.plot(X, Y, color='red', linestyle=':')
plt.plot(X, Y, 'g:', label='theory')
plt.plot(X2, Y2+.5, 'd', label='experiment')


plt.xlabel('X label')
plt.ylabel('Y label')
plt.legend()
plt.title(r'My title $E = mc^2 = \hbar\omega$')
plt.text(2, 0, r'$\alpha + \beta/\gamma^2$')

plt.ylim(-2, 2)
plt.xlim(0, 7)

plt.savefig('my_graph.pdf')
plt.savefig('my_graph.svg')
```

```python
x = np.logspace(0, 2, 51)

y = x**0.453 + x

plt.loglog(x, y)
plt.grid(which='both')
```

```python
plt.subplot(2, 1, 1)
plt.plot(X, Y)

plt.subplot(2, 1, 2)
plt.plot(X2, Y2, 'o')
```

```python
x = np.linspace(-3, 3)
X, Y = np.meshgrid(x, x)

R_square = X**2 + Y**2
Z = np.exp(-R_square)

plt.imshow(Z)
plt.colorbar()
```

## Object oriented

```python
from matplotlib.pyplot import figure

X = np.linspace(0, 10, 200)
X2 = np.linspace(0, 10, 20)

Y = np.sin(X)
Y2 = np.sin(X2)


f = figure()
ax1, ax2 = f.subplots(2, 1)

ax1.plot(X, Y, 'g:')
ax2.plot(X2, Y2, 'o')
```
